```python
# CELL 5 — Extract drug mentions and temporal words from discharge notes
import pandas as pd
import re

mt = pd.read_csv('medical-nlp/medical-nlp-master/data/mtsamples.csv')
discharge = mt[mt['medical_specialty'].str.strip() == 'Discharge Summary'].reset_index(drop=True)

# Common drug names to look for
drug_keywords = [
    'aspirin', 'ibuprofen', 'acetaminophen', 'tylenol', 'warfarin',
    'heparin', 'insulin', 'metformin', 'lisinopril', 'metoprolol',
    'atorvastatin', 'amoxicillin', 'vancomycin', 'ceftriaxone',
    'morphine', 'oxycodone', 'prednisone', 'bactrim', 'lasix',
    'furosemide', 'digoxin', 'coumadin', 'lovenox', 'plavix',
    'zofran', 'ondansetron', 'dilaudid', 'ativan', 'lorazepam',
    'fentanyl', 'midazolam', 'propofol', 'labetalol', 'hydralazine'
]

# Temporal words that signal time relationships
temporal_keywords = [
    'before', 'after', 'prior to', 'following', 'subsequently',
    'initially', 'then', 'later', 'first', 'second', 'third',
    'postoperative', 'on admission', 'at discharge', 'next day',
    'day 1', 'day 2', 'day 3', 'on day', 'postop',
    'upon arrival', 'thereafter', 'eventually', 'previously'
]

results = []

for i, row in discharge.iterrows():
    text = str(row['transcription']).lower()

    # Find drugs mentioned
    drugs_found = [d for d in drug_keywords if d in text]

    # Find temporal words mentioned
    temporal_found = [t for t in temporal_keywords if t in text]

    # Only keep notes that have BOTH drugs AND temporal words
    if len(drugs_found) >= 1 and len(temporal_found) >= 2:
        results.append({
            'note_id': i,
            'description': row['description'],
            'drugs_found': ', '.join(drugs_found),
            'temporal_words': ', '.join(temporal_found),
            'drug_count': len(drugs_found),
            'temporal_count': len(temporal_found),
            'text': row['transcription']
        })

df_results = pd.DataFrame(results)
print(f"Total discharge notes         : {len(discharge)}")
print(f"Notes with drugs + time words : {len(df_results)}")
print()
print("--- NOTES FOUND (showing top 10) ---")
print(df_results[['note_id','description','drugs_found','temporal_words']].head(10).to_string(index=False))
```

```python
# CELL 6 — Generate temporal QA benchmark questions from discharge notes
import pandas as pd
import re

mt = pd.read_csv('medical-nlp/medical-nlp-master/data/mtsamples.csv')
discharge = mt[mt['medical_specialty'].str.strip() == 'Discharge Summary'].reset_index(drop=True)

drug_keywords = [
    'aspirin', 'ibuprofen', 'acetaminophen', 'tylenol', 'warfarin',
    'heparin', 'insulin', 'metformin', 'lisinopril', 'metoprolol',
    'atorvastatin', 'amoxicillin', 'vancomycin', 'ceftriaxone',
    'morphine', 'oxycodone', 'prednisone', 'bactrim', 'lasix',
    'furosemide', 'digoxin', 'coumadin', 'lovenox', 'plavix',
    'zofran', 'ondansetron', 'dilaudid', 'ativan', 'lorazepam',
    'fentanyl', 'midazolam', 'propofol', 'labetalol', 'hydralazine'
]

# Patterns that signal BEFORE relationship
before_patterns = [
    r'(\w+)\s+(?:was given|was started|administered|received)\s+.*?before\s+(\w+)',
    r'prior to\s+(\w+).*?(\w+)\s+(?:was|were)',
    r'(\w+)\s+initially.*?then\s+(\w+)',
    r'first.*?(\w+).*?then.*?(\w+)',
    r'(\w+)\s+on admission.*?(\w+)\s+at discharge',
]

# Patterns that signal AFTER relationship
after_patterns = [
    r'(\w+)\s+(?:was given|was started|administered)\s+.*?after\s+(\w+)',
    r'following\s+(\w+).*?(\w+)\s+(?:was|were|started)',
    r'subsequently.*?(\w+)',
    r'(\w+)\s+postoperative.*?(\w+)',
]

questions = []

for i, row in discharge.iterrows():
    text = str(row['transcription'])
    text_lower = text.lower()

    drugs_in_note = [d for d in drug_keywords if d in text_lower]

    if len(drugs_in_note) < 1:
        continue

    # Split into sentences
    sentences = re.split(r'[.!?]', text)

    for sent in sentences:
        sent_lower = sent.lower().strip()
        if len(sent_lower) < 20:
            continue

        drugs_in_sent = [d for d in drugs_in_note if d in sent_lower]

        # Generate BEFORE questions
        if ('before' in sent_lower or 'prior to' in sent_lower or
            'initially' in sent_lower) and len(drugs_in_sent) >= 1:

            drug = drugs_in_sent[0].upper()
            question = f"Was {drug} administered before the clinical event described in this sentence?"
            questions.append({
                'note_id': i,
                'sentence': sent.strip(),
                'drug': drug,
                'question': question,
                'relation_type': 'BEFORE',
                'answer': 'YES',
                'full_note_snippet': text[:300]
            })

        # Generate AFTER questions
        if ('after' in sent_lower or 'following' in sent_lower or
            'subsequently' in sent_lower) and len(drugs_in_sent) >= 1:

            drug = drugs_in_sent[0].upper()
            question = f"Was {drug} administered after the clinical event described in this sentence?"
            questions.append({
                'note_id': i,
                'sentence': sent.strip(),
                'drug': drug,
                'question': question,
                'relation_type': 'AFTER',
                'answer': 'YES',
                'full_note_snippet': text[:300]
            })

        # Generate SIMULTANEOUS questions (drugs given together)
        if len(drugs_in_sent) >= 2:
            d1 = drugs_in_sent[0].upper()
            d2 = drugs_in_sent[1].upper()
            question = f"Were {d1} and {d2} mentioned together in the same clinical context?"
            questions.append({
                'note_id': i,
                'sentence': sent.strip(),
                'drug': f"{d1} + {d2}",
                'question': question,
                'relation_type': 'SIMULTANEOUS',
                'answer': 'YES',
                'full_note_snippet': text[:300]
            })

# Save to CSV
df_questions = pd.DataFrame(questions)
df_questions.to_csv('temporal_benchmark.csv', index=False)

print(f"✅ Benchmark dataset created!")
print(f"Total questions generated : {len(df_questions)}")
print()
print("Questions by type:")
print(df_questions['relation_type'].value_counts().to_string())
print()
print("--- SAMPLE QUESTIONS (first 10) ---")
for idx, row in df_questions.head(10).iterrows():
    print(f"\nQ{idx+1} [{row['relation_type']}]")
    print(f"  Sentence : {row['sentence'][:120]}")
    print(f"  Drug     : {row['drug']}")
    print(f"  Question : {row['question']}")
    print(f"  Answer   : {row['answer']}")
```

```python
# CELL 7 — Expand benchmark using more specialties + smarter extraction
import pandas as pd
import re

mt = pd.read_csv('medical-nlp/medical-nlp-master/data/mtsamples.csv')

# Use these specialties — they all contain drug timelines
good_specialties = [
    'Discharge Summary',
    'General Medicine',
    'Emergency Room Reports',
    'Cardiovascular / Pulmonary',
    'Neurology',
    'Nephrology',
    'Hematology - Oncology',
    'Gastroenterology',
    'SOAP / Chart / Progress Notes',
    'Pediatrics - Neonatal'
]

# Expanded drug list
drug_keywords = [
    'aspirin', 'ibuprofen', 'acetaminophen', 'tylenol', 'warfarin',
    'heparin', 'insulin', 'metformin', 'lisinopril', 'metoprolol',
    'atorvastatin', 'amoxicillin', 'vancomycin', 'ceftriaxone',
    'morphine', 'oxycodone', 'prednisone', 'bactrim', 'lasix',
    'furosemide', 'digoxin', 'coumadin', 'lovenox', 'plavix',
    'zofran', 'ondansetron', 'dilaudid', 'ativan', 'lorazepam',
    'fentanyl', 'midazolam', 'propofol', 'labetalol', 'hydralazine',
    'penicillin', 'clindamycin', 'azithromycin', 'ciprofloxacin',
    'dexamethasone', 'methylprednisolone', 'albuterol', 'nitroglycerin',
    'amlodipine', 'carvedilol', 'spironolactone', 'clopidogrel',
    'enoxaparin', 'apixaban', 'rivaroxaban', 'metronidazole',
    'fluconazole', 'acyclovir', 'tamoxifen', 'carboplatin',
    'dopamine', 'epinephrine', 'norepinephrine', 'dobutamine',
    'phenytoin', 'diazepam', 'haloperidol', 'quetiapine',
    'omeprazole', 'pantoprazole', 'enalapril', 'ramipril',
    'simvastatin', 'rosuvastatin', 'levothyroxine', 'amiodarone'
]

# Temporal signal words
before_signals = ['before', 'prior to', 'initially', 'first',
                  'on admission', 'upon arrival', 'previously',
                  'preoperative', 'pre-op', 'earlier']

after_signals  = ['after', 'following', 'subsequently', 'then',
                  'postoperative', 'postop', 'post-op', 'later',
                  'thereafter', 'at discharge', 'next day',
                  'on day 1', 'on day 2', 'on day 3', 'day 2',
                  'day 3', 'second day', 'third day']

simul_signals  = ['and', 'along with', 'combination', 'together',
                  'additionally', 'also', 'as well as', 'concomitant']

questions = []
question_id = 0

# Filter to good specialties
subset = mt[mt['medical_specialty'].str.strip().isin(good_specialties)].reset_index(drop=True)
print(f"Notes being processed: {len(subset)}")

for i, row in subset.iterrows():
    text     = str(row['transcription'])
    text_low = text.lower()
    specialty = str(row['medical_specialty']).strip()

    drugs_in_note = [d for d in drug_keywords if d in text_low]
    if not drugs_in_note:
        continue

    # Split into sentences
    sentences = re.split(r'(?<=[.!?])\s+', text)

    for sent in sentences:
        sent_low  = sent.lower().strip()
        if len(sent_low) < 25:
            continue

        drugs_in_sent = [d for d in drugs_in_note if d in sent_low]
        if not drugs_in_sent:
            continue

        before_found = [w for w in before_signals if w in sent_low]
        after_found  = [w for w in after_signals  if w in sent_low]

        # BEFORE question
        if before_found:
            question_id += 1
            questions.append({
                'qid'          : question_id,
                'note_id'      : i,
                'specialty'    : specialty,
                'sentence'     : sent.strip(),
                'drug'         : drugs_in_sent[0].upper(),
                'trigger_word' : before_found[0],
                'question'     : f"Was {drugs_in_sent[0].upper()} given BEFORE the event described?",
                'relation'     : 'BEFORE',
                'answer'       : 'YES'
            })

        # AFTER question
        if after_found:
            question_id += 1
            questions.append({
                'qid'          : question_id,
                'note_id'      : i,
                'specialty'    : specialty,
                'sentence'     : sent.strip(),
                'drug'         : drugs_in_sent[0].upper(),
                'trigger_word' : after_found[0],
                'question'     : f"Was {drugs_in_sent[0].upper()} given AFTER the event described?",
                'relation'     : 'AFTER',
                'answer'       : 'YES'
            })

        # SIMULTANEOUS question
        if len(drugs_in_sent) >= 2:
            question_id += 1
            d1 = drugs_in_sent[0].upper()
            d2 = drugs_in_sent[1].upper()
            questions.append({
                'qid'          : question_id,
                'note_id'      : i,
                'specialty'    : specialty,
                'sentence'     : sent.strip(),
                'drug'         : f"{d1} + {d2}",
                'trigger_word' : 'co-mention',
                'question'     : f"Were {d1} and {d2} given in the same clinical context?",
                'relation'     : 'SIMULTANEOUS',
                'answer'       : 'YES'
            })

df_q = pd.DataFrame(questions).drop_duplicates(subset=['sentence','relation'])
df_q.to_csv('temporal_benchmark_v2.csv', index=False)

print(f"\n✅ Expanded benchmark saved!")
print(f"Total questions   : {len(df_q)}")
print(f"\nBy relation type:")
print(df_q['relation'].value_counts().to_string())
print(f"\nBy medical specialty:")
print(df_q['specialty'].value_counts().to_string())
print(f"\n--- 5 SAMPLE QUESTIONS ---")
for _, row in df_q.sample(5, random_state=42).iterrows():
    print(f"\n[{row['relation']}] Trigger: '{row['trigger_word']}'")
    print(f"  Drug    : {row['drug']}")
    print(f"  Sentence: {row['sentence'][:130]}")
    print(f"  Question: {row['question']}")
    print(f"  Answer  : {row['answer']}")
```

```python
# CELL 11 — Run ClinicalBERT on your benchmark dataset
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import numpy as np
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Load your benchmark
df = pd.read_csv('/content/drive/MyDrive/ChronoClinician_Research/temporal_benchmark_v2.csv')
print(f"✅ Benchmark loaded: {len(df)} questions")

# Load ClinicalBERT
print("\nLoading ClinicalBERT... (takes 1-2 minutes)")
model_name = "medicalai/ClinicalBERT"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=3)
model.eval()
print("✅ ClinicalBERT loaded!")

# Label mapping
label2id = {'BEFORE': 0, 'AFTER': 1, 'SIMULTANEOUS': 2}
id2label = {0: 'BEFORE', 1: 'AFTER', 2: 'SIMULTANEOUS'}

# Run inference on each question
print("\nRunning ClinicalBERT on all 472 questions...")
predictions = []

for idx, row in tqdm(df.iterrows(), total=len(df)):
    # Format input: [question] + [sentence]
    input_text = f"{row['question']} [SEP] {row['sentence']}"

    # Tokenize
    inputs = tokenizer(
        input_text,
        return_tensors='pt',
        truncation=True,
        max_length=512,
        padding=True
    )

    # Get prediction
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        pred_id = torch.argmax(logits, dim=1).item()
        pred_label = id2label[pred_id]
        confidence = torch.softmax(logits, dim=1).max().item()

    predictions.append({
        'qid': row['qid'],
        'true_label': row['relation'],
        'predicted_label': pred_label,
        'confidence': round(confidence, 3),
        'correct': pred_label == row['relation']
    })

# Results
results_df = pd.DataFrame(predictions)
df_eval = df.merge(results_df, on='qid')

# Overall accuracy
overall_acc = results_df['correct'].mean()
print(f"\n{'='*50}")
print("CLINICALBERT RESULTS ON YOUR BENCHMARK")
print(f"{'='*50}")
print(f"Overall Accuracy : {overall_acc:.1%}")
print()

# Per-class accuracy
print("Accuracy by Relation Type:")
for rel in ['BEFORE', 'AFTER', 'SIMULTANEOUS']:
    subset = results_df[results_df['true_label'] == rel]
    acc = subset['correct'].mean()
    print(f"  {rel:<15} {acc:.1%}  ({len(subset)} questions)")

print()

# Confusion matrix style breakdown
print("Prediction Distribution:")
print(results_df['predicted_label'].value_counts().to_string())

# Save results
results_df.to_csv('/content/drive/MyDrive/ChronoClinician_Research/clinicalbert_results.csv', index=False)
print()
print("✅ Results saved to Google Drive!")
```

```python
# CELL 12 — ClinicalBERT Zero-Shot baseline (fair evaluation)
import pandas as pd
import torch
from transformers import pipeline
import warnings
warnings.filterwarnings('ignore')

# Load benchmark
df = pd.read_csv('/content/drive/MyDrive/ChronoClinician_Research/temporal_benchmark_v2.csv')
print(f"✅ Benchmark loaded: {len(df)} questions")

# Use zero-shot classification pipeline
# This is the FAIR way to evaluate a model not fine-tuned on your task
print("\nLoading zero-shot classifier...")
classifier = pipeline(
    "zero-shot-classification",
    model="typeform/distilbert-base-uncased-mnli",
    device=-1  # CPU
)
print("✅ Zero-shot classifier loaded!")

# Candidate labels — the 3 temporal relations
candidate_labels = ["BEFORE", "AFTER", "SIMULTANEOUS"]

print(f"\nRunning zero-shot classification on {len(df)} questions...")
print("(This takes ~5 minutes, you will see progress below)\n")

predictions = []
correct = 0

for idx, row in df.iterrows():
    # Input = sentence + question combined
    input_text = f"{row['sentence']} {row['question']}"

    # Truncate if too long
    input_text = input_text[:500]

    try:
        result = classifier(input_text, candidate_labels)
        pred_label = result['labels'][0]  # Highest scoring label
        confidence = result['scores'][0]
        is_correct = (pred_label == row['relation'])
        if is_correct:
            correct += 1
    except:
        pred_label = "SIMULTANEOUS"
        confidence = 0.0
        is_correct = False

    predictions.append({
        'qid'            : row['qid'],
        'true_label'     : row['relation'],
        'predicted_label': pred_label,
        'confidence'     : round(confidence, 3),
        'correct'        : is_correct,
        'sentence'       : row['sentence'][:100],
        'drug'           : row['drug']
    })

    # Print progress every 50 questions
    if (idx + 1) % 50 == 0:
        running_acc = correct / (idx + 1)
        print(f"  Processed {idx+1}/472 | Running accuracy: {running_acc:.1%}")

# Final results
results_df = pd.DataFrame(predictions)
overall_acc = results_df['correct'].mean()

print(f"\n{'='*50}")
print("ZERO-SHOT BASELINE RESULTS")
print(f"{'='*50}")
print(f"Overall Accuracy : {overall_acc:.1%}")
print()
print("Accuracy by Relation Type:")
for rel in ['BEFORE', 'AFTER', 'SIMULTANEOUS']:
    subset = results_df[results_df['true_label'] == rel]
    acc = subset['correct'].mean()
    count = len(subset)
    print(f"  {rel:<15} {acc:.1%}  ({count} questions)")

print()
print("What the model predicted (distribution):")
print(results_df['predicted_label'].value_counts().to_string())

# Show some correct examples
print("\n--- 3 QUESTIONS IT GOT RIGHT ---")
correct_examples = results_df[results_df['correct'] == True].head(3)
for _, row in correct_examples.iterrows():
    print(f"\n  Drug     : {row['drug']}")
    print(f"  Sentence : {row['sentence'][:100]}")
    print(f"  True     : {row['true_label']}")
    print(f"  Predicted: {row['predicted_label']}")
    print(f"  Confidence: {row['confidence']:.1%}")

# Show some wrong examples
print("\n--- 3 QUESTIONS IT GOT WRONG ---")
wrong_examples = results_df[results_df['correct'] == False].head(3)
for _, row in wrong_examples.iterrows():
    print(f"\n  Drug     : {row['drug']}")
    print(f"  Sentence : {row['sentence'][:100]}")
    print(f"  True     : {row['true_label']}")
    print(f"  Predicted: {row['predicted_label']}")
    print(f"  Confidence: {row['confidence']:.1%}")

# Save
results_df.to_csv(
    '/content/drive/MyDrive/ChronoClinician_Research/zeroshot_baseline_results.csv',
    index=False
)
print(f"\n✅ Results saved to Google Drive!")
```

```python
# CELL 15 — Fine-tune BioBERT on your benchmark (Your Proposed Model)
import pandas as pd
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import warnings
warnings.filterwarnings('ignore')

# ── 1. Load benchmark ─────────────────────────────────
df = pd.read_csv('/content/drive/MyDrive/ChronoClinician_Research/temporal_benchmark_v2.csv')
print(f"✅ Benchmark loaded: {len(df)} questions")

# ── 2. Label encoding ─────────────────────────────────
label2id = {'BEFORE': 0, 'AFTER': 1, 'SIMULTANEOUS': 2}
id2label  = {0: 'BEFORE', 1: 'AFTER', 2: 'SIMULTANEOUS'}
df['label'] = df['relation'].map(label2id)

# ── 3. Train / Test split (80/20) ─────────────────────
train_df, test_df = train_test_split(
    df, test_size=0.2, random_state=42, stratify=df['label']
)
print(f"Train samples : {len(train_df)}")
print(f"Test  samples : {len(test_df)}")
print()

# ── 4. Dataset class ──────────────────────────────────
class TemporalDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_len=256):
        self.data      = dataframe.reset_index(drop=True)
        self.tokenizer = tokenizer
        self.max_len   = max_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        # Input = question + separator + clinical sentence
        text = f"{row['question']} [SEP] {row['sentence']}"
        enc  = self.tokenizer(
            text,
            truncation   = True,
            max_length   = self.max_len,
            padding      = 'max_length',
            return_tensors = 'pt'
        )
        return {
            'input_ids'      : enc['input_ids'].squeeze(),
            'attention_mask' : enc['attention_mask'].squeeze(),
            'label'          : torch.tensor(row['label'], dtype=torch.long)
        }

# ── 5. Load BioBERT tokenizer + model ─────────────────
print("Loading BioBERT... (1-2 minutes)")
MODEL_NAME = "dmis-lab/biobert-base-cased-v1.2"
tokenizer  = AutoTokenizer.from_pretrained(MODEL_NAME)
model      = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME, num_labels=3
)
print("✅ BioBERT loaded!")

# ── 6. Create dataloaders ─────────────────────────────
train_dataset = TemporalDataset(train_df, tokenizer)
test_dataset  = TemporalDataset(test_df,  tokenizer)

train_loader  = DataLoader(train_dataset, batch_size=8, shuffle=True)
test_loader   = DataLoader(test_dataset,  batch_size=8, shuffle=False)

# ── 7. Training setup ─────────────────────────────────
device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"\nRunning on: {device}")
model     = model.to(device)
optimizer = AdamW(model.parameters(), lr=2e-5)

# ── 8. Training loop (3 epochs) ───────────────────────
print("\n" + "="*50)
print("TRAINING YOUR MODEL")
print("="*50)

EPOCHS = 3
for epoch in range(EPOCHS):
    model.train()
    total_loss = 0
    correct    = 0
    total      = 0

    for batch_idx, batch in enumerate(train_loader):
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels         = batch['label'].to(device)

        optimizer.zero_grad()
        outputs = model(
            input_ids      = input_ids,
            attention_mask = attention_mask,
            labels         = labels
        )

        loss = outputs.loss
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        preds       = torch.argmax(outputs.logits, dim=1)
        correct    += (preds == labels).sum().item()
        total      += labels.size(0)

        if (batch_idx + 1) % 10 == 0:
            print(f"  Epoch {epoch+1} | Batch {batch_idx+1}/{len(train_loader)} "
                  f"| Loss: {total_loss/(batch_idx+1):.3f} "
                  f"| Train Acc: {correct/total:.1%}")

    print(f"\n✅ Epoch {epoch+1} complete | "
          f"Avg Loss: {total_loss/len(train_loader):.3f} | "
          f"Train Acc: {correct/total:.1%}\n")

# ── 9. Evaluation ─────────────────────────────────────
print("="*50)
print("EVALUATING ON TEST SET")
print("="*50)

model.eval()
all_preds  = []
all_labels = []

with torch.no_grad():
    for batch in test_loader:
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels         = batch['label'].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        preds   = torch.argmax(outputs.logits, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# ── 10. Results ───────────────────────────────────────
all_preds  = np.array(all_preds)
all_labels = np.array(all_labels)
overall_acc = accuracy_score(all_labels, all_preds)

print(f"\nOverall Accuracy : {overall_acc:.1%}")
print()
print("Per-Class Results:")
for class_id, class_name in id2label.items():
    mask     = all_labels == class_id
    class_acc = accuracy_score(all_labels[mask], all_preds[mask])
    count    = mask.sum()
    print(f"  {class_name:<15} {class_acc:.1%}  ({count} questions)")

print()
print("Full Classification Report:")
print(classification_report(
    all_labels, all_preds,
    target_names=['BEFORE','AFTER','SIMULTANEOUS']
))

# ── 11. Save results ──────────────────────────────────
pred_df = pd.DataFrame({
    'true_label'      : [id2label[l] for l in all_labels],
    'predicted_label' : [id2label[p] for p in all_preds],
    'correct'         : all_labels == all_preds
})
pred_df.to_csv(
    '/content/drive/MyDrive/ChronoClinician_Research/biobert_mamba_results.csv',
    index=False
)
print("✅ Results saved to Google Drive!")
```

```python
# CELL 16 — Harder evaluation: remove trigger words (anti-cheat test)
import pandas as pd
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('/content/drive/MyDrive/ChronoClinician_Research/temporal_benchmark_v2.csv')

# List of trigger words to REMOVE from sentences
trigger_words = [
    'before', 'prior to', 'initially', 'first', 'on admission',
    'upon arrival', 'previously', 'preoperative', 'pre-op', 'earlier',
    'after', 'following', 'subsequently', 'then', 'postoperative',
    'postop', 'post-op', 'later', 'thereafter', 'at discharge',
    'next day', 'second day', 'third day', 'day 1', 'day 2', 'day 3'
]

import re

def remove_triggers(text):
    text_lower = text.lower()
    for tw in trigger_words:
        text_lower = text_lower.replace(tw, '[TIME]')
    return text_lower

# Create harder version of benchmark
df['sentence_hard'] = df['sentence'].apply(remove_triggers)
df['question_hard'] = df['question'].apply(remove_triggers)

print("Original sentence example:")
print(df['sentence'].iloc[5][:150])
print()
print("After removing trigger words:")
print(df['sentence_hard'].iloc[5][:150])
print()

# Label encoding
label2id = {'BEFORE': 0, 'AFTER': 1, 'SIMULTANEOUS': 2}
id2label  = {0: 'BEFORE', 1: 'AFTER', 2: 'SIMULTANEOUS'}
df['label'] = df['relation'].map(label2id)

# Train/test split
train_df, test_df = train_test_split(
    df, test_size=0.2, random_state=42, stratify=df['label']
)
print(f"Train: {len(train_df)} | Test: {len(test_df)}")

# Dataset using HARD sentences
class HardTemporalDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_len=256):
        self.data      = dataframe.reset_index(drop=True)
        self.tokenizer = tokenizer
        self.max_len   = max_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row  = self.data.iloc[idx]
        text = f"{row['question_hard']} [SEP] {row['sentence_hard']}"
        enc  = self.tokenizer(
            text,
            truncation     = True,
            max_length     = self.max_len,
            padding        = 'max_length',
            return_tensors = 'pt'
        )
        return {
            'input_ids'      : enc['input_ids'].squeeze(),
            'attention_mask' : enc['attention_mask'].squeeze(),
            'label'          : torch.tensor(row['label'], dtype=torch.long)
        }

# Load BioBERT fresh
print("\nLoading BioBERT...")
MODEL_NAME = "dmis-lab/biobert-base-cased-v1.2"
tokenizer  = AutoTokenizer.from_pretrained(MODEL_NAME)
model      = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME, num_labels=3
)

device    = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model     = model.to(device)
optimizer = AdamW(model.parameters(), lr=2e-5)

train_dataset = HardTemporalDataset(train_df, tokenizer)
test_dataset  = HardTemporalDataset(test_df,  tokenizer)
train_loader  = DataLoader(train_dataset, batch_size=8, shuffle=True)
test_loader   = DataLoader(test_dataset,  batch_size=8, shuffle=False)

# Train
print("\nTraining on HARD benchmark (trigger words removed)...")
EPOCHS = 3
for epoch in range(EPOCHS):
    model.train()
    total_loss = 0
    correct = 0
    total   = 0
    for batch in train_loader:
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels         = batch['label'].to(device)
        optimizer.zero_grad()
        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            labels=labels
        )
        outputs.loss.backward()
        optimizer.step()
        total_loss += outputs.loss.item()
        preds       = torch.argmax(outputs.logits, dim=1)
        correct    += (preds == labels).sum().item()
        total      += labels.size(0)
    print(f"Epoch {epoch+1} | Loss: {total_loss/len(train_loader):.3f} | Train Acc: {correct/total:.1%}")

# Evaluate
print("\nEvaluating on HARD test set...")
model.eval()
all_preds  = []
all_labels = []
with torch.no_grad():
    for batch in test_loader:
        input_ids      = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels         = batch['label'].to(device)
        outputs        = model(input_ids=input_ids, attention_mask=attention_mask)
        preds          = torch.argmax(outputs.logits, dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

all_preds  = np.array(all_preds)
all_labels = np.array(all_labels)

print(f"\n{'='*55}")
print("HARD BENCHMARK RESULTS (trigger words removed)")
print(f"{'='*55}")
print(f"Overall Accuracy: {accuracy_score(all_labels, all_preds):.1%}")
print()
print("Per-Class Results:")
for cid, cname in id2label.items():
    mask = all_labels == cid
    acc  = accuracy_score(all_labels[mask], all_preds[mask])
    print(f"  {cname:<15} {acc:.1%}  ({mask.sum()} questions)")
print()
print(classification_report(
    all_labels, all_preds,
    target_names=['BEFORE','AFTER','SIMULTANEOUS']
))

# Save
pd.DataFrame({
    'true' : [id2label[l] for l in all_labels],
    'pred' : [id2label[p] for p in all_preds],
    'correct': all_labels == all_preds
}).to_csv(
    '/content/drive/MyDrive/ChronoClinician_Research/biobert_hard_results.csv',
    index=False
)
print("✅ Hard benchmark results saved!")
```
